```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./ml-100k/u.data',sep='\t')
df.head()
```

```python
df.columns = ['user_id','item_id','rating','timestamp']
df.head()
```

```python
ratings = pd.read_csv('./ml-100k/u.item',sep="\|",encoding = "ISO-8859-1",header=None)
ratings.head()
```

```python
ratings = ratings[[0,1]]
ratings.columns = ['item_id','title']
ratings.head()
```

```python
df = df.merge(ratings,on='item_id')
df.head()
```

```python
df1 = df.groupby('title').mean()['rating']
df1 = pd.DataFrame(df1)
df1
```

```python
df1['no. of rating'] = pd.DataFrame(df.groupby('title').count()['rating'])
df1
```

# data visualisation

```python
plt.figure(figsize=(10,6))
plt.hist(df1['no. of rating'],bins=70)
plt.show()
```

```python
plt.hist(df1['rating'],bins=70)
plt.show()
```

```python
sns.jointplot(x='rating',y='no. of rating',data=df1,alpha=0.7)
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
df1.sort_values('no. of rating',ascending=False).head()
```

```python
starwars_user_rating = moviemat['Star Wars (1977)']
starwars_user_rating
```

```python
corr_starwars = moviemat.corrwith(starwars_user_rating)
corr_starwars.dropna(inplace=True)
corr_starwars
```

```python
similar_to_starwars = pd.DataFrame(corr_starwars,columns=['correlation'])
similar_to_starwars.sort_values('correlation',ascending=False)
```

```python
# if a movie has less number of rating and all those user rated starwars and this movie 
# then they have perfect correlation but this will affect recommendation
# we want to set some threshold of no. of rating given to the movie and their correlation with starwars
similar_to_starwars=similar_to_starwars.join(df1['no. of rating'],on='title')
similar_to_starwars
```

```python
similar_to_starwars[similar_to_starwars['no. of rating']>100].sort_values('correlation',ascending=False)
```

## Predict movie

```python
def predict_movie(movie_name): 
    movie_user_ratings = moviemat[movie_name]
    corr_with_movie = moviemat.corrwith(movie_user_ratings)
    corr_with_movie.dropna(inplace=True)
    
    similar_to_movie = pd.DataFrame(corr_with_movie,columns=['correlation'])
    similar_to_movie = similar_to_movie.join(df1['no. of rating'],on='title')
    
    predictions = similar_to_movie[similar_to_movie['no. of rating']>100].sort_values('correlation',ascending=False)
    return predictions
```

```python
predictions = predict_movie('Titanic (1997)')
predictions.head()
```
