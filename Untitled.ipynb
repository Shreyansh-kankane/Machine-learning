```python
import pandas as pd
from keras.datasets import mnist
import matplotlib.pyplot as plt
```

```python
(X_train,y_train),(X_test,y_test) = mnist.load_data()
```

```python
X_test.shape
```

```python
plt.imshow(X_test[90],cmap='gray')
```

```python
y_test[90]
```

```python
X = X_test.reshape(-1,28*28)
y = y_test
X.shape, y.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
# standard scaling i,e mean around 0 and std around 1
sc = StandardScaler()
X_ = sc.fit_transform(X)
X_.shape
```

```python
X_.mean(),X_.std() 
```

# Preprocessing

```python
plt.imshow(X_[90].reshape(28,28),cmap='gray')
plt.show()
```

# Sklearn PCA

```python
from sklearn.decomposition import PCA
```

```python
# convert from 784 to 2 dimension 
pca = PCA(n_components=2)
```

```python
z_pca = pca.fit_transform(X_) 
z_pca.shape
```

```python
# variance of these reduced dimension
pca.explained_variance_
```

# Custom PCA

```python
import numpy as np
```

```python
# compute covariance matrix
covar_matrix = np.dot(X_.T,X_)
covar_matrix.shape
```

```python
# compute eigen vector using SVD (singular value decomposition)
```

```python
from numpy.linalg import svd
```

```python
U,S,V = svd(covar_matrix)
U.shape
```

```python
Ured = U[:,:2]
Ured.shape
```

```python
# Projection of data on new axis 
```

```python
z = np.dot(X_,Ured)
z.shape
```

```python
z #new x with 2 dimension
```

# Visualise dataset

```python
new_dataset = np.hstack((z,y.reshape(-1,1))) #horizontal stacking with y
```

```python
df = pd.DataFrame(new_dataset,columns=['PC1','PC2','label'])
df.head()
```

```python
import seaborn as sns
plt.figure(figsize=(15,15))
fg = sns.FacetGrid(df,hue='label',height=10)
fg.map(plt.scatter,"PC1","PC2")
fg.add_legend()
plt.show()
```

# PCA with all dimension

```python
pca = PCA()
```

```python
Z_pca = pca.fit_transform(X_)
Z_pca.shape
```

```python
pca.explained_variance_
```

```python
cum_var_explained = np.cumsum(pca.explained_variance_ratio_)
cum_var_explained
```

```python
plt.figure(figsize=(8,6))
plt.plot(cum_var_explained)
plt.grid()
plt.xlabel("n_components")
plt.ylabel("cummulative explained variance")
plt.show()
```
